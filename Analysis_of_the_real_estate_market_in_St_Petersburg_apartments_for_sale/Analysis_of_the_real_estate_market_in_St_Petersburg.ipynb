## Предобработка данных

### Корректировка названий столбцов

```python
# Создадим список из названий столбцов подлежащих замене и выполним ее.
data.rename(columns = {'cityCenters_nearest': 'city_сenters_nearest', 'parks_around3000': 'parks_around_3000', 'ponds_around3000': 'ponds_around_3000'}, inplace = True)
```

### Поиск пропусков

Отфильтруем столбцы с пропусками по возрастанию. Посмотрим на их отношение к общему количеству строк, чтобы понять насколько они критичны.

```python
data.isna().sum().sort_values()
```

```python
data.isnull().sum().sort_values()/data.shape[0]
```

Пропуски в столбцах `locality_name` и `floors_total` малочисленны. Их можно удалить, но лучше оставить или заменить на подходящие значения. Возможно в дальнейшем они помогут найти нам какие-либо инсайты.

### Обработка столбца  'locality_name'

```python
# Выведем количество пропусков в столбце 'locality_name'
data['locality_name'].isna().sum()
```

```python
# Изучим содержимое столбца locality_name.
data['locality_name'].unique()
```

В списке значений втречается слово 'поселок' в двух вариантах написания с 'е' и 'ё'. Изменим это.

```python
data['locality_name'] = data['locality_name'].str.replace('посёлок','поселок')
```

```python
# Также приведем к общему формату вариант написания 'поселок городского типа'. Заменим его на 'поселок'.
data['locality_name'] = data['locality_name'].str.replace('поселок городского типа','поселок')
```

```python
# Проверим произошла ли замена.
data['locality_name'].unique()
```

Возможно при заполнении карточки объявления пользователь надеялся на автоматическое заполнение графы с населенным пунктом, так как указал улицу и номер дом. Скорее всего пропуски вызваны человеческим фактором. 

Так как геоданных для восстановления названия населенного пункта у нас нет, а квартира без точного адреса не имеет никакой ценности мы удалим строки с подобными пропусками.

```python
data.dropna(subset = ['locality_name'], inplace = True)
```

```python
# Убедимся что пропусков не осталось.
data['locality_name'].isna().sum()
```

### Обработка столбца  'floors_total'

```python
# Выведем количество пропусков в столбце 'floors_total'
data['floors_total'].isna().sum()
```

```python
# Рассмотрим уникальные значения столбца 'floors_total'
data['floors_total'].unique()
```

Возможно отсутствие количества этажей в доме объясняется человеческим фактором. Пользователь при заполнении анкеты забыл указать этажность дома или не захотел. Так как этаж мог быть последним, он решил не вносить данные. 

Мы можем заменить отсутствие данных на нетипичное значение, чтобы в дальнейшем при расчетах учитывать эти данные.

```python
# Выведем на экран строчку с квартирой на 60-м этаже.
data.query('floors_total == "60"')
```

Квартира находится в Кронштаде. Высота домов в городе ограничена [15 метрами](https://www.fontanka.ru/2021/06/04/69951491/). Скорее всего это ошибка при заполнении данных. Отфильтруем эту строчку.

```python
# Заполним пропуски значением -1.
data['floors_total'].fillna(-1, inplace=True)
```

```python
# Убедимся что пропусков не осталось.
data['floors_total'].isna().sum()
```

```python
data = data[data['floors_total'] != 60]
```

```python
data
```

```python
# Изменим тип данных столбца 'floors_total' на int так как количество этажей это целое число.
data['floors_total'] = data['floors_total'].astype('int64')
data.dtypes
```

```python
# Проверим работу кода.
data['floors_total'].unique()
```

### Обработка столбца  'living_area'

```python
#Выведем количество пропусков в столбце 'living_area'
data['living_area'].isna().sum()
```

```python
#Изучим содержание строчек датафрейма со значением NaN в столбце 'living_area'
data.query('living_area == "NaN"').head()
```

Иногда риэлторы в своей практике используют [публикацию фиктивных объявлений](https://www.avito.ru/blog/re-case-aska?utm_medium=cpc&avito_campaign_id=62633207&utm_source=yandex&utm_campaign=web_yandex_listers_b2b_realty_all_search_08062021_any_generic_b2b_all-content-aon&utm_content=text&utm_term=%D1%80%D1%8B%D0%BD%D0%BE%D0%BA%20%D0%BD%D0%B5%D0%B4%D0%B2%D0%B8%D0%B6%D0%B8%D0%BC%D0%BE%D1%81%D1%82%D0%B8&yclid=761487201140267744). Часто они указывают низкую цену или публикуют фотографии с очень хорошим ремонтом. В телефонном разговоре сообщают что  квартиру уже продали, но есть другие варианты. Возможно это тот самый случай.

```python
# Расчитаем медианное значение по общему объему жилой площали в соответствии с количеством комнат в квартире. Заполним пропуски.
data['living_area'] = data['living_area'].fillna(data.groupby('rooms')['living_area'].transform('median'))
```

```python
# Убедимся что пропусков не осталось.
data['living_area'].isna().sum()
```

### Обработка столбца  'kitchen_area'

В перечне значений замечены квартиры с высотой потолков более 10 м. Рассмотрим эти цифры ближе. 

```python
# Для удобства сохраним срез в новую переменную 'data_check'
data_check = data.query('ceiling_height >= 10')
```

Возможно квартиры с потолками 10.3 и 14.0 м это многоуровневые пентхаусы. А квартиры со значением 20 м и выше это результат ошибки при вводе данных. Пользователь не внес знак между целым и сотыми. Исправим это разделив высокие значения на 10. 

```python
# Напишем функцию для замены аномальных значений
def ceiling_height_anomalies(values):
    if values >= 20.0:
        res = values / 10
        return res
    return values

# Проверим ее работу
print(ceiling_height_anomalies(9))
print(ceiling_height_anomalies(20.10))
print(ceiling_height_anomalies(100))
```

```python
# Применим функцию к столбцу перезаписав его.
data['ceiling_height'] = data['ceiling_height'].apply(ceiling_height_anomalies)
```

```python
# Заполним медианным значением пропуски в столбце 'ceiling_height' сгрупировав данные по столбцу 'floors_total'
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('floors_total')['ceiling_height'].transform('median'))
```

### Категоризация квартир по этажности дома. 

```python
# Создадим функцию для категоризации значений: варианты — первый, последний, другой
def building_floor_category(row):
    if row['floor'] == row['floors_total'] == 1:
        return 'первый'
    elif (row['floor'] == 1) and (row['floor'] != row['floors_total']):
        return 'первый'
    elif 1 < row['floor'] < row['floors_total']:
        return 'другой'
    elif (1 < row['floor']) and (row['floor'] == row['floors_total']):
        return 'последний'
# Помним о том что среди данных столбца 'floors_total' встречаются нетипичные значения -1. Пропишем еще один вариант категории 'неопределен'
    else:
        return 'неопределен'
```

```python
# Применим функцию к столбцу и выведем получившиеся значения на экран
data['floor_cathegory'] = data.apply(building_floor_category, axis=1)
data[['floors_total', 'floor', 'floor_cathegory']].head()
```

Дополнительно изучим распределение объявлений по месяцам и средней стоимости по годам. Для точности анализа будем использовать еще необработанные данные

```python
# Сохраним исходные данные в новую переменную df
df = pd.read_csv('/datasets/real_estate_data.csv', sep='\t')

# Добавим в данные столбцы с месяцем и годом публикации объявлений
df['month_of_publication'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year_of_publication'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

```python
# Создадим словарь с перечнем действий, которые мы планируем совершить над данными
agg_func = {'month_of_publication': 'count', 'last_price': 'mean'}

# Сгруппируем данные по году и месяцу публикации.
distribution_of_publications_years = df.groupby(['year_of_publication']).agg(agg_func).reset_index()
distribution_of_publications_years
```

В 2014 году в базе Яндекс.Недвижимости по Санкт-Петербургу находилось всего 137 объявлений. Стоимость лотов  в них была выше средней по рынку. В последующие годы базу пополнили объявления массового продавца и значение графика поползло вниз. 

```python
distribution_of_publications_months = df.pivot_table(index='year_of_publication', columns='month_of_publication', values='last_price', aggfunc ='mean').reset_index()
distribution_of_publications_months
```

В данных за 2019 год хранятся объявления только за 5 месяцев. Несмотря на это средняя стоимость лотов выше значения 2018 года. * Нестандартное соотношение цены, площади квартиры и района ее расположения
* Задублированные объявления
* Отсутствие важной информации в объявлении влияющей на оценку стоимости. Например общей площади квартиры или комбинации из нескольких параметров.
